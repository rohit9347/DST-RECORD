```python
import pandas as pd

# Sample DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [24, 30, 22, 35],
    'City': ['New York', 'London', 'Paris', 'Tokyo']
}

df = pd.DataFrame(data)

# -----------------------------
# Selecting a column
ages = df['Age']

# Filtering rows
above_25 = df[df['Age'] > 25]

# iloc (integer-based) vs loc (label-based)
row_0 = df.iloc[0]                 # First row
specific_val = df.loc[0, 'City']    # Value at row label 0, column 'City'

# -----------------------------
# Display results
print("Ages column:\n", ages, "\n")
print("Age > 25:\n", above_25, "\n")
print("Row 0 using iloc:\n", row_0, "\n")
print("Specific value using loc:", specific_val)
```

```python
# MATPLOTLIB
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Create dataset
data = {
    'StudyHours': [1, 2, 3, 4, 5, 6, 7, 8],
    'ExamScore': [35, 40, 50, 55, 65, 70, 78, 85]
}

df = pd.DataFrame(data)

# -----------------------------

# 1️ Scatter Plot
plt.scatter(df['StudyHours'], df['ExamScore'])
plt.xlabel("Study Hours")
plt.ylabel("Exam Score")
plt.title("Study Hours vs Exam Score (Scatter Plot)")
plt.show()
```

```python
#SCIKIT

import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# STEP 2: CREATE DATASET
data = {
    'Student': ['S1','S2','S3','S4','S5','S6','S7','S8','S9','S10'],
    'Classes_Attended': [30, 35, 40, 45, 50, 55, 60, 65, 70, 75],
    'Internal_Marks': [35, 38, 42, 46, 50, 55, 60, 65, 70, 75]
}

df = pd.DataFrame(data)
print("\nDataset:\n", df)

# STEP 3: FEATURES & TARGET
X = df[['Classes_Attended']]
y = df['Internal_Marks']

# STEP 4: TRAIN-TEST SPLIT
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=7
)

# STEP 5: TRAIN MODEL
model = LinearRegression()
model.fit(X_train, y_train)

# STEP 6: PREDICTION
predictions = model.predict(X_test)

# STEP 7: EVALUATION
mse = mean_squared_error(y_test, predictions)
slope = model.coef_[0]
intercept = model.intercept_

print("\nModel Performance")
print("------------------")
print("Mean Squared Error (MSE):", round(mse, 3))
print("Marks per Class Attended:", round(slope, 3))
print("Base Marks (Intercept):", round(intercept, 3))
print(f"Regression Equation: Marks = {round(slope,3)} × Classes + {round(intercept,3)}")

# STEP 8: VISUALIZATION
plt.figure()
plt.scatter(X, y, label="Actual Data")
plt.plot(X, model.predict(X), label="Regression Line")
plt.xlabel("Classes Attended")
plt.ylabel("Internal Marks")
plt.title("Classes Attended vs Internal Marks")
plt.legend()
plt.show()
```
